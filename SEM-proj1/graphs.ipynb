```python
import pandas as pd

def stats(data):
    print(data[['DHT','DS18B20','LM35DZ','thermistor','humidity']].describe())
    print(data[['DHT','DS18B20','LM35DZ','thermistor']].values.mean(),"-",data[['DHT','DS18B20','LM35DZ','thermistor']].values.std())
```

```python
df = pd.read_csv('measurements/short1.txt',names=['date','DHT','DS18B20','LM35DZ','thermistor','humidity'])
df.loc[:, df.columns != 'humidity'].plot(x='date', rot=45,ylabel='˚C')
df[['date','humidity']].plot(x='date', rot=45,ylabel='%')
stats(df)
```
